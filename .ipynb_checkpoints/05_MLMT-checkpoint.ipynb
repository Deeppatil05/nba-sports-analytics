```python
# Phase 5: Machine Learning Model Training
# Purpose: Build predictive models for sports betting

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from pathlib import Path
from datetime import datetime

# ML Libraries
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                             f1_score, roc_auc_score, confusion_matrix,
                             classification_report, roc_curve)
from sklearn.preprocessing import StandardScaler
import xgboost as xgb

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 100)

print("="*70)
print("PHASE 5: MACHINE LEARNING MODEL TRAINING")
print("="*70)
print(f"\nStart time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")

# Paths
PROCESSED_DIR = Path('data/processed')
REPORTS_DIR = Path('reports')
REPORTS_DIR.mkdir(exist_ok=True)

print("üìÇ Loading engineered datasets...")
train = pd.read_csv(PROCESSED_DIR / 'train_set.csv')
val = pd.read_csv(PROCESSED_DIR / 'val_set.csv')
test = pd.read_csv(PROCESSED_DIR / 'test_set.csv')

print(f"‚úÖ Train: {len(train):,} games (2006-2014)")
print(f"‚úÖ Val:   {len(val):,} games (2015-2016)")
print(f"‚úÖ Test:  {len(test):,} games (2017)")
print(f"\nüìä Total: {len(train) + len(val) + len(test):,} games")


# Define feature columns and target
print("\nüéØ PREPARING FEATURES & TARGET")
print("="*70)

# Exclude non-feature columns
exclude_cols = [
    'GAME_DATE_EST', 'GAME_ID', 'GAME_STATUS_TEXT', 'SEASON', 
    'TEAM_ID_home', 'TEAM_ID_away', 'HOME_TEAM_ID', 'VISITOR_TEAM_ID',
    'GAME_DATE', 'MATCHUP', 'HOME_TEAM', 'AWAY_TEAM', 'YEAR', 'MONTH', 'DAY_OF_WEEK',
    # Targets (outcomes we're predicting)
    'PTS_home', 'PTS_away', 'HOME_TEAM_WINS', 'SPREAD_HOME_COVER', 
    'TOTAL_OVER', 'ML_HOME_WIN', 'ML_AWAY_WIN',
    # Post-game data (would be leakage)
    'ACTUAL_MARGIN', 'ACTUAL_TOTAL',
    # Betting lines (already encoded as features)
    'HOME_MONEYLINE', 'AWAY_MONEYLINE', 'HOME_SPREAD', 'AWAY_SPREAD', 'TOTAL_LINE',
    'HOME_WIN_PROB', 'AWAY_WIN_PROB',
    # In-game stats (not available before game)
    'FG_PCT_home', 'FG_PCT_away', 'FT_PCT_home', 'FT_PCT_away',
    'FG3_PCT_home', 'FG3_PCT_away', 'AST_home', 'AST_away', 
    'REB_home', 'REB_away'
]

# Select numeric feature columns
feature_cols = [col for col in train.columns if col not in exclude_cols]
feature_cols = [col for col in feature_cols if train[col].dtype in ['int64', 'float64']]

print(f"‚úÖ Selected {len(feature_cols)} features for modeling")

# Target variable
target = 'SPREAD_HOME_COVER'

# Prepare datasets
X_train = train[feature_cols].fillna(0)
y_train = train[target]

X_val = val[feature_cols].fillna(0)
y_val = val[target]

X_test = test[feature_cols].fillna(0)
y_test = test[target]

print(f"\nüìä Target: {target}")
print(f"   Train positive rate: {y_train.mean()*100:.1f}%")
print(f"   Val positive rate:   {y_val.mean()*100:.1f}%")
print(f"   Test positive rate:  {y_test.mean()*100:.1f}%")

print(f"\nüìã Sample features:")
print(feature_cols[:10])
print("   ...")
print(f"   (showing 10 of {len(feature_cols)})")


# Feature scaling (important for Logistic Regression)
print("\n‚öñÔ∏è SCALING FEATURES")
print("="*70)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# Convert back to DataFrame (for feature names)
X_train_scaled = pd.DataFrame(X_train_scaled, columns=feature_cols, index=X_train.index)
X_val_scaled = pd.DataFrame(X_val_scaled, columns=feature_cols, index=X_val.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=feature_cols, index=X_test.index)

print("‚úÖ Features scaled (mean=0, std=1)")
print(f"   Train mean: {X_train_scaled.mean().mean():.6f}")
print(f"   Train std:  {X_train_scaled.std().mean():.6f}")


# Baseline: Always predict home team covers
print("\nüìä BASELINE MODEL: Always Bet Home Team")
print("="*70)

baseline_pred = np.ones(len(y_test))  # Always predict 1 (home covers)

baseline_acc = accuracy_score(y_test, baseline_pred)
baseline_precision = precision_score(y_test, baseline_pred)
baseline_recall = recall_score(y_test, baseline_pred)

print(f"\nBaseline Performance (Always Bet Home):")
print(f"   Accuracy:  {baseline_acc*100:.2f}%")
print(f"   Precision: {baseline_precision*100:.2f}%")
print(f"   Recall:    {baseline_recall*100:.2f}%")

print(f"\nüí° This is our benchmark to beat!")
print(f"   Any model below {baseline_acc*100:.1f}% accuracy is worse than random!")


# Model 1: Logistic Regression (simple, interpretable)
print("\nü§ñ MODEL 1: LOGISTIC REGRESSION")
print("="*70)

print("Training Logistic Regression...")
lr_model = LogisticRegression(max_iter=1000, random_state=42, n_jobs=-1)
lr_model.fit(X_train_scaled, y_train)

# Predictions
lr_train_pred = lr_model.predict(X_train_scaled)
lr_val_pred = lr_model.predict(X_val_scaled)
lr_test_pred = lr_model.predict(X_test_scaled)

# Probabilities (for ROC)
lr_train_proba = lr_model.predict_proba(X_train_scaled)[:, 1]
lr_val_proba = lr_model.predict_proba(X_val_scaled)[:, 1]
lr_test_proba = lr_model.predict_proba(X_test_scaled)[:, 1]

# Performance metrics
print("\nüìä Logistic Regression Performance:")
print("\n" + " "*15 + "Train    Val      Test")
print(f"Accuracy:     {accuracy_score(y_train, lr_train_pred)*100:6.2f}% {accuracy_score(y_val, lr_val_pred)*100:6.2f}% {accuracy_score(y_test, lr_test_pred)*100:6.2f}%")
print(f"Precision:    {precision_score(y_train, lr_train_pred)*100:6.2f}% {precision_score(y_val, lr_val_pred)*100:6.2f}% {precision_score(y_test, lr_test_pred)*100:6.2f}%")
print(f"Recall:       {recall_score(y_train, lr_train_pred)*100:6.2f}% {recall_score(y_val, lr_val_pred)*100:6.2f}% {recall_score(y_test, lr_test_pred)*100:6.2f}%")
print(f"F1 Score:     {f1_score(y_train, lr_train_pred)*100:6.2f}% {f1_score(y_val, lr_val_pred)*100:6.2f}% {f1_score(y_test, lr_test_pred)*100:6.2f}%")
print(f"ROC AUC:      {roc_auc_score(y_train, lr_train_proba):6.4f}  {roc_auc_score(y_val, lr_val_proba):6.4f}  {roc_auc_score(y_test, lr_test_proba):6.4f}")

# Top coefficients
feature_coefs = pd.DataFrame({
    'Feature': feature_cols,
    'Coefficient': lr_model.coef_[0]
}).sort_values('Coefficient', key=abs, ascending=False)

print("\nüîç Top 10 Most Important Features (by coefficient):")
print(feature_coefs.head(10).to_string(index=False))

print("\n‚úÖ Logistic Regression trained!")


# Model 2: Random Forest (handles non-linearity)
print("\nüå≤ MODEL 2: RANDOM FOREST")
print("="*70)

print("Training Random Forest (100 trees)...")
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=15,
    min_samples_split=50,
    min_samples_leaf=20,
    random_state=42,
    n_jobs=-1
)
rf_model.fit(X_train, y_train)  # No scaling needed for tree models

# Predictions
rf_train_pred = rf_model.predict(X_train)
rf_val_pred = rf_model.predict(X_val)
rf_test_pred = rf_model.predict(X_test)

# Probabilities
rf_train_proba = rf_model.predict_proba(X_train)[:, 1]
rf_val_proba = rf_model.predict_proba(X_val)[:, 1]
rf_test_proba = rf_model.predict_proba(X_test)[:, 1]

# Performance
print("\nüìä Random Forest Performance:")
print("\n" + " "*15 + "Train    Val      Test")
print(f"Accuracy:     {accuracy_score(y_train, rf_train_pred)*100:6.2f}% {accuracy_score(y_val, rf_val_pred)*100:6.2f}% {accuracy_score(y_test, rf_test_pred)*100:6.2f}%")
print(f"Precision:    {precision_score(y_train, rf_train_pred)*100:6.2f}% {precision_score(y_val, rf_val_pred)*100:6.2f}% {precision_score(y_test, rf_test_pred)*100:6.2f}%")
print(f"Recall:       {recall_score(y_train, rf_train_pred)*100:6.2f}% {recall_score(y_val, rf_val_pred)*100:6.2f}% {recall_score(y_test, rf_test_pred)*100:6.2f}%")
print(f"F1 Score:     {f1_score(y_train, rf_train_pred)*100:6.2f}% {f1_score(y_val, rf_val_pred)*100:6.2f}% {f1_score(y_test, rf_test_pred)*100:6.2f}%")
print(f"ROC AUC:      {roc_auc_score(y_train, rf_train_proba):6.4f}  {roc_auc_score(y_val, rf_val_proba):6.4f}  {roc_auc_score(y_test, rf_test_proba):6.4f}")

# Feature importance
rf_importance = pd.DataFrame({
    'Feature': feature_cols,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nüîç Top 10 Most Important Features:")
print(rf_importance.head(10).to_string(index=False))

print("\n‚úÖ Random Forest trained!")


# Model 3: XGBoost (state-of-the-art gradient boosting)
print("\nüöÄ MODEL 3: XGBOOST")
print("="*70)

print("Training XGBoost...")
xgb_model = xgb.XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1,
    eval_metric='logloss'
)

# Train with early stopping on validation set
xgb_model.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    verbose=False
)

# Predictions
xgb_train_pred = xgb_model.predict(X_train)
xgb_val_pred = xgb_model.predict(X_val)
xgb_test_pred = xgb_model.predict(X_test)

# Probabilities
xgb_train_proba = xgb_model.predict_proba(X_train)[:, 1]
xgb_val_proba = xgb_model.predict_proba(X_val)[:, 1]
xgb_test_proba = xgb_model.predict_proba(X_test)[:, 1]

# Performance
print("\nüìä XGBoost Performance:")
print("\n" + " "*15 + "Train    Val      Test")
print(f"Accuracy:     {accuracy_score(y_train, xgb_train_pred)*100:6.2f}% {accuracy_score(y_val, xgb_val_pred)*100:6.2f}% {accuracy_score(y_test, xgb_test_pred)*100:6.2f}%")
print(f"Precision:    {precision_score(y_train, xgb_train_pred)*100:6.2f}% {precision_score(y_val, xgb_val_pred)*100:6.2f}% {precision_score(y_test, xgb_test_pred)*100:6.2f}%")
print(f"Recall:       {recall_score(y_train, xgb_train_pred)*100:6.2f}% {recall_score(y_val, xgb_val_pred)*100:6.2f}% {recall_score(y_test, xgb_test_pred)*100:6.2f}%")
print(f"F1 Score:     {f1_score(y_train, xgb_train_pred)*100:6.2f}% {f1_score(y_val, xgb_val_pred)*100:6.2f}% {f1_score(y_test, xgb_test_pred)*100:6.2f}%")
print(f"ROC AUC:      {roc_auc_score(y_train, xgb_train_proba):6.4f}  {roc_auc_score(y_val, xgb_val_proba):6.4f}  {roc_auc_score(y_test, xgb_test_proba):6.4f}")

# Feature importance
xgb_importance = pd.DataFrame({
    'Feature': feature_cols,
    'Importance': xgb_model.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nüîç Top 10 Most Important Features:")
print(xgb_importance.head(10).to_string(index=False))

print("\n‚úÖ XGBoost trained!")


# Compare all models
print("\nüìä MODEL COMPARISON SUMMARY")
print("="*70)

comparison = pd.DataFrame({
    'Model': ['Baseline', 'Logistic Regression', 'Random Forest', 'XGBoost'],
    'Test_Accuracy': [
        baseline_acc * 100,
        accuracy_score(y_test, lr_test_pred) * 100,
        accuracy_score(y_test, rf_test_pred) * 100,
        accuracy_score(y_test, xgb_test_pred) * 100
    ],
    'Test_Precision': [
        baseline_precision * 100,
        precision_score(y_test, lr_test_pred) * 100,
        precision_score(y_test, rf_test_pred) * 100,
        precision_score(y_test, xgb_test_pred) * 100
    ],
    'Test_ROC_AUC': [
        0.50,  # Baseline has no probability
        roc_auc_score(y_test, lr_test_proba),
        roc_auc_score(y_test, rf_test_proba),
        roc_auc_score(y_test, xgb_test_proba)
    ]
})

print("\n" + comparison.to_string(index=False))

# Highlight best model
best_model_idx = comparison['Test_ROC_AUC'].idxmax()
best_model = comparison.loc[best_model_idx, 'Model']
best_auc = comparison.loc[best_model_idx, 'Test_ROC_AUC']

print(f"\nüèÜ BEST MODEL: {best_model} (ROC AUC: {best_auc:.4f})")


# Visualization: Confusion Matrices
print("\nüìä CONFUSION MATRICES")
print("="*70)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

models = [
    ('Logistic Regression', lr_test_pred),
    ('Random Forest', rf_test_pred),
    ('XGBoost', xgb_test_pred)
]

for idx, (name, predictions) in enumerate(models):
    cm = confusion_matrix(y_test, predictions)
    
    # Plot
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[idx],
                cbar=False, square=True, annot_kws={'size': 16})
    axes[idx].set_title(f'{name}\nTest Set Confusion Matrix', 
                        fontsize=13, fontweight='bold')
    axes[idx].set_ylabel('True Label', fontsize=11)
    axes[idx].set_xlabel('Predicted Label', fontsize=11)
    axes[idx].set_xticklabels(['Away Covers', 'Home Covers'])
    axes[idx].set_yticklabels(['Away Covers', 'Home Covers'])
    
    # Add accuracy in corner
    acc = accuracy_score(y_test, predictions)
    axes[idx].text(0.5, 2.3, f'Accuracy: {acc*100:.1f}%', 
                   fontsize=10, ha='center',
                   bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.savefig('reports/10_confusion_matrices.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Saved: reports/10_confusion_matrices.png")


# ROC Curves
print("\nüìà ROC CURVES")
print("="*70)

plt.figure(figsize=(10, 8))

models_proba = [
    ('Logistic Regression', lr_test_proba),
    ('Random Forest', rf_test_proba),
    ('XGBoost', xgb_test_proba)
]

colors = ['blue', 'green', 'red']

for (name, proba), color in zip(models_proba, colors):
    fpr, tpr, _ = roc_curve(y_test, proba)
    auc = roc_auc_score(y_test, proba)
    
    plt.plot(fpr, tpr, color=color, linewidth=2, 
             label=f'{name} (AUC = {auc:.3f})')

# Diagonal (random classifier)
plt.plot([0, 1], [0, 1], 'k--', linewidth=2, label='Random (AUC = 0.500)')

plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('ROC Curves - Model Comparison\n(Test Set)', 
          fontsize=14, fontweight='bold')
plt.legend(loc='lower right', fontsize=11)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('reports/11_roc_curves.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Saved: reports/11_roc_curves.png")


# Compare feature importance across models
print("\nüîç FEATURE IMPORTANCE COMPARISON")
print("="*70)

# Get top 15 features from XGBoost (best model)
top_features = xgb_importance.head(15)['Feature'].tolist()

# Compare across models
importance_comparison = pd.DataFrame({
    'Feature': top_features,
    'Logistic_Coef': [abs(feature_coefs[feature_coefs['Feature'] == f]['Coefficient'].values[0]) 
                      for f in top_features],
    'RandomForest': [rf_importance[rf_importance['Feature'] == f]['Importance'].values[0] 
                     for f in top_features],
    'XGBoost': [xgb_importance[xgb_importance['Feature'] == f]['Importance'].values[0] 
                for f in top_features]
})

# Normalize to 0-100 scale
for col in ['Logistic_Coef', 'RandomForest', 'XGBoost']:
    importance_comparison[col] = (
        100 * importance_comparison[col] / importance_comparison[col].max()
    )

# Visualize
fig, ax = plt.subplots(figsize=(12, 10))

x = np.arange(len(top_features))
width = 0.25

ax.barh(x - width, importance_comparison['Logistic_Coef'], width, 
        label='Logistic Regression', alpha=0.8, color='blue')
ax.barh(x, importance_comparison['RandomForest'], width, 
        label='Random Forest', alpha=0.8, color='green')
ax.barh(x + width, importance_comparison['XGBoost'], width, 
        label='XGBoost', alpha=0.8, color='red')

ax.set_yticks(x)
ax.set_yticklabels(top_features)
ax.invert_yaxis()
ax.set_xlabel('Relative Importance (Normalized to 100)', fontsize=12)
ax.set_title('Top 15 Features - Importance Across Models', 
             fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(alpha=0.3, axis='x')

plt.tight_layout()
plt.savefig('reports/12_feature_importance_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

print("‚úÖ Saved: reports/12_feature_importance_comparison.png")

print("\nüîç Top 5 Most Consistently Important Features:")
avg_importance = importance_comparison[['Logistic_Coef', 'RandomForest', 'XGBoost']].mean(axis=1)
importance_comparison['Avg_Importance'] = avg_importance
print(importance_comparison.nlargest(5, 'Avg_Importance')[['Feature', 'Avg_Importance']].to_string(index=False))


# BETTING ROI SIMULATION
print("\nüí∞ BETTING ROI SIMULATION (XGBoost Model)")
print("="*70)

# Add predictions to test set
test_results = test.copy()
test_results['XGB_PRED'] = xgb_test_pred
test_results['XGB_PROB'] = xgb_test_proba
test_results['XGB_CONFIDENCE'] = np.abs(xgb_test_proba - 0.5)  # Distance from 50%

# Strategy 1: Bet on every game model predicts
print("\nüìä STRATEGY 1: Bet Every Game")
print("-" * 70)

total_bets = len(test_results)
wins = (test_results['XGB_PRED'] == test_results['SPREAD_HOME_COVER']).sum()
win_rate = wins / total_bets * 100

# Standard betting: -110 odds (need to win 52.4% to break even)
# Win: +$100, Lose: -$110
profit_per_win = 100
loss_per_loss = 110

total_profit = (wins * profit_per_win) - ((total_bets - wins) * loss_per_loss)
roi = (total_profit / (total_bets * loss_per_loss)) * 100

print(f"Total bets:    {total_bets:,}")
print(f"Wins:          {wins:,}")
print(f"Losses:        {total_bets - wins:,}")
print(f"Win rate:      {win_rate:.2f}%")
print(f"Total profit:  ${total_profit:,.2f}")
print(f"ROI:           {roi:.2f}%")

if win_rate > 52.4:
    print(f"‚úÖ WIN RATE > 52.4% - This strategy is PROFITABLE!")
else:
    print(f"‚ùå WIN RATE < 52.4% - Need 52.4% to break even at -110 odds")

# Strategy 2: Only bet high confidence predictions
print("\nüìä STRATEGY 2: Bet Only High Confidence (>60% probability)")
print("-" * 70)

high_conf = test_results[test_results['XGB_PROB'] > 0.6]
total_bets_hc = len(high_conf)

if total_bets_hc > 0:
    wins_hc = (high_conf['XGB_PRED'] == high_conf['SPREAD_HOME_COVER']).sum()
    win_rate_hc = wins_hc / total_bets_hc * 100
    
    total_profit_hc = (wins_hc * profit_per_win) - ((total_bets_hc - wins_hc) * loss_per_loss)
    roi_hc = (total_profit_hc / (total_bets_hc * loss_per_loss)) * 100
    
    print(f"Total bets:    {total_bets_hc:,}")
    print(f"Wins:          {wins_hc:,}")
    print(f"Losses:        {total_bets_hc - wins_hc:,}")
    print(f"Win rate:      {win_rate_hc:.2f}%")
    print(f"Total profit:  ${total_profit_hc:,.2f}")
    print(f"ROI:           {roi_hc:.2f}%")
    
    if win_rate_hc > 52.4:
        print(f"‚úÖ WIN RATE > 52.4% - High confidence betting is PROFITABLE!")
    else:
        print(f"‚ùå Still not profitable - model needs improvement")
else:
    print("‚ö†Ô∏è  No predictions with >60% confidence")

# Strategy 3: Bet only on home underdogs (exploiting the 63.7% anomaly)
print("\nüìä STRATEGY 3: Bet Home Underdogs Only")
print("-" * 70)

home_dogs = test_results[test_results['HOME_UNDERDOG'] == 1]
total_bets_hd = len(home_dogs)

if total_bets_hd > 0:
    # Using model predictions
    wins_hd = (home_dogs['XGB_PRED'] == home_dogs['SPREAD_HOME_COVER']).sum()
    win_rate_hd = wins_hd / total_bets_hd * 100
    
    total_profit_hd = (wins_hd * profit_per_win) - ((total_bets_hd - wins_hd) * loss_per_loss)
    roi_hd = (total_profit_hd / (total_bets_hd * loss_per_loss)) * 100
    
    print(f"Total bets:    {total_bets_hd:,}")
    print(f"Wins:          {wins_hd:,}")
    print(f"Losses:        {total_bets_hd - wins_hd:,}")
    print(f"Win rate:      {win_rate_hd:.2f}%")
    print(f"Total profit:  ${total_profit_hd:,.2f}")
    print(f"ROI:           {roi_hd:.2f}%")
    
    # Compare to always betting home underdogs (no model)
    wins_baseline = home_dogs['SPREAD_HOME_COVER'].sum()
    win_rate_baseline = wins_baseline / total_bets_hd * 100
    profit_baseline = (wins_baseline * profit_per_win) - ((total_bets_hd - wins_baseline) * loss_per_loss)
    
    print(f"\n   vs. Baseline (always bet home dogs):")
    print(f"   Win rate: {win_rate_baseline:.2f}%")
    print(f"   Profit:   ${profit_baseline:,.2f}")
    print(f"   Model improvement: {win_rate_hd - win_rate_baseline:+.2f} percentage points")


# Analyze ROI by confidence level
print("\nüìä ROI BY CONFIDENCE LEVEL")
print("="*70)

# Create confidence bins
confidence_bins = [0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 1.0]
test_results['CONF_BIN'] = pd.cut(test_results['XGB_PROB'], bins=confidence_bins)

roi_by_conf = []

for bin_range in test_results['CONF_BIN'].cat.categories:
    subset = test_results[test_results['CONF_BIN'] == bin_range]
    
    if len(subset) > 0:
        wins = (subset['XGB_PRED'] == subset['SPREAD_HOME_COVER']).sum()
        total = len(subset)
        win_rate = wins / total * 100
        profit = (wins * 100) - ((total - wins) * 110)
        roi = (profit / (total * 110)) * 100
        
        roi_by_conf.append({
            'Confidence': str(bin_range),
            'Bets': total,
            'Wins': wins,
            'Win_Rate': win_rate,
            'Profit': profit,
            'ROI': roi
        })

roi_df = pd.DataFrame(roi_by_conf)
print("\n" + roi_df.to_string(index=False))

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Win rate by confidence
axes[0].bar(range(len(roi_df)), roi_df['Win_Rate'], alpha=0.7, color='blue')
axes[0].axhline(y=52.4, color='red', linestyle='--', linewidth=2, label='Break-even (52.4%)')
axes[0].set_xticks(range(len(roi_df)))
axes[0].set_xticklabels(roi_df['Confidence'], rotation=45, ha='right')
axes[0].set_xlabel('Confidence Level', fontsize=12)
axes[0].set_ylabel('Win Rate (%)', fontsize=12)
axes[0].set_title('Win Rate by Model Confidence', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].grid(alpha=0.3, axis='y')

# ROI by confidence
colors = ['red' if x < 0 else 'green' for x in roi_df['ROI']]
axes[1].bar(range(len(roi_df)), roi_df['ROI'], alpha=0.7, color=colors)
axes[1].axhline(y=0, color='black', linestyle='-', linewidth=1)
axes[1].set_xticks(range(len(roi_df)))
axes[1].set_xticklabels(roi_df['Confidence'], rotation=45, ha='right')
axes[1].set_xlabel('Confidence Level', fontsize=12)
axes[1].set_ylabel('ROI (%)', fontsize=12)
axes[1].set_title('ROI by Model Confidence', fontsize=14, fontweight='bold')
axes[1].grid(alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('reports/13_roi_by_confidence.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n‚úÖ Saved: reports/13_roi_by_confidence.png")

# Find optimal confidence threshold
profitable_bins = roi_df[roi_df['ROI'] > 0]
if len(profitable_bins) > 0:
    best_conf = profitable_bins.loc[profitable_bins['ROI'].idxmax()]
    print(f"\nüèÜ OPTIMAL BETTING STRATEGY:")
    print(f"   Bet only when confidence in: {best_conf['Confidence']}")
    print(f"   Expected win rate: {best_conf['Win_Rate']:.1f}%")
    print(f"   Expected ROI: {best_conf['ROI']:.1f}%")
    print(f"   Number of bets: {best_conf['Bets']}")
else:
    print("\n‚ö†Ô∏è  No confidence level shows positive ROI in test set")



# Save trained models and results
print("\nüíæ SAVING MODELS & RESULTS")
print("="*70)

import joblib

# Save models
joblib.dump(lr_model, PROCESSED_DIR / 'logistic_regression_model.pkl')
joblib.dump(rf_model, PROCESSED_DIR / 'random_forest_model.pkl')
joblib.dump(xgb_model, PROCESSED_DIR / 'xgboost_model.pkl')
joblib.dump(scaler, PROCESSED_DIR / 'feature_scaler.pkl')

print("‚úÖ Saved trained models:")
print("   - logistic_regression_model.pkl")
print("   - random_forest_model.pkl")
print("   - xgboost_model.pkl")
print("   - feature_scaler.pkl")

# Save test predictions
test_results[['GAME_ID', 'GAME_DATE', 'MATCHUP', 'HOME_SPREAD', 
              'SPREAD_HOME_COVER', 'XGB_PRED', 'XGB_PROB']].to_csv(
    PROCESSED_DIR / 'test_predictions.csv', index=False
)
print("‚úÖ Saved test_predictions.csv")

# Save model comparison
comparison.to_csv(PROCESSED_DIR / 'model_comparison.csv', index=False)
print("‚úÖ Saved model_comparison.csv")

# Save feature importance
xgb_importance.to_csv(PROCESSED_DIR / 'xgboost_feature_importance.csv', index=False)
print("‚úÖ Saved xgboost_feature_importance.csv")


# Final summary
print("\n" + "="*70)
print("‚úÖ PHASE 5 COMPLETE - MACHINE LEARNING MODELS")
print("="*70)

print("\nü§ñ MODELS TRAINED:")
print("   1. Logistic Regression (baseline)")
print("   2. Random Forest (ensemble)")
print("   3. XGBoost (best performance)")

print("\nüìä BEST MODEL: XGBoost")
xgb_test_acc = accuracy_score(y_test, xgb_test_pred)
xgb_test_auc = roc_auc_score(y_test, xgb_test_proba)
print(f"   Test Accuracy: {xgb_test_acc*100:.2f}%")
print(f"   Test ROC AUC:  {xgb_test_auc:.4f}")
print(f"   Improvement over baseline: {(xgb_test_acc - baseline_acc)*100:+.2f} percentage points")

print("\nüí∞ BETTING PERFORMANCE:")
print(f"   Strategy 1 (bet all): {win_rate:.1f}% win rate, ${total_profit:,.0f} profit, {roi:.1f}% ROI")
if 'win_rate_hc' in locals():
    print(f"   Strategy 2 (high conf): {win_rate_hc:.1f}% win rate, ${total_profit_hc:,.0f} profit, {roi_hc:.1f}% ROI")
if 'win_rate_hd' in locals():
    print(f"   Strategy 3 (home dogs): {win_rate_hd:.1f}% win rate, ${total_profit_hd:,.0f} profit, {roi_hd:.1f}% ROI")

print("\nüîç TOP 5 FEATURES:")
for idx, row in xgb_importance.head(5).iterrows():
    print(f"   {idx+1}. {row['Feature']}: {row['Importance']:.4f}")

print("\nüìÅ SAVED FILES:")
print("   ‚Ä¢ Models: 4 .pkl files")
print("   ‚Ä¢ Results: test_predictions.csv")
print("   ‚Ä¢ Analysis: model_comparison.csv, feature_importance.csv")
print("   ‚Ä¢ Visualizations: 4 charts in reports/")

print("\nüéØ NEXT STEPS:")
print("   ‚Üí Create final presentation/report")
print("   ‚Üí Document findings for portfolio")
print("   ‚Üí Deploy model (optional)")
print("   ‚Üí Backtest on more recent seasons")

print(f"\nüìÇ All outputs saved to: {PROCESSED_DIR} and reports/")
print("\n" + "="*70)
print(f"Model training completed: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print("="*70)

print("\nüéâ CONGRATULATIONS! You've built a complete sports betting ML system!")
print("\nüí¨ Next: Create final portfolio presentation? (Y/N)")


    
```
